```python
# dependencies 

import pandas as pd
import json
import datetime
```

```python
# compile all csv files into one dict; separated by FA, FG and bbref batter and pitcher 

data = {}
for i in range(2006,2017,1):
    fa = pd.read_csv(str(i)+"FA.csv")
    fa = fa[fa["PLAYER"]!="PLAYER"]
    fa = fa[fa["DOLLARS"]=="Minor Lg"]
    fbat = pd.read_csv(str(i+1)+" bleaders.csv")
    fpitch = pd.read_csv(str(i+1)+" pleaders.csv")
    bbat = pd.read_csv(str(i+1)+" bWAR batters.csv")
    bpitch = pd.read_csv(str(i+1)+" bWAR pitchers.csv")
    data[i] = [fa,{"FG":[fbat,fpitch], "BBREF":[bbat, bpitch]}]
data
```

```python
# create dict with all players for each year - both their bWAR and fWAR as well as their PA and IP if applicable

a = datetime.datetime.now()
players = {}
for e in data:
    cur = e+1
    players[cur] = [{}]
    fadata = data[e][0]
    seadata = data[e][1]
    for index1, row1 in fadata.iterrows():
        fa = row1["PLAYER"]
        for site in seadata:
            k = 0
            for f in seadata[site]:
                for index2, row2 in f.iterrows():
                    if fa[:fa.find(" ")] in row2["Name"] and fa[fa.find(" ")+1:] in row2["Name"] or fa == row2["Name"]:
                        if fa not in players[cur][0]:
                            players[cur][0][fa] = {"fWAR":0.0, "bWAR":0.0}
                        war = row2["WAR"]
                        if site == "FG":
                            players[cur][0][fa]["fWAR"] += war
                            if k == 1:
                                players[cur][0][fa]["IP"] = row2["IP"]
                            else:
                                players[cur][0][fa]["PA"] = row2["PA"]
                        else:
                            if row2["WAR"] == row2["WAR"]:
                                players[cur][0][fa]["bWAR"] += war
                k+=1
b = datetime.datetime.now()
print(b-a)
```

```python
# for each year, add all the IP, PA, bWAR, and fWAR and store in dict

totals = {}
for e in players:
    totals[e] = {"Total IP": 0, "Total PA": 0, "Total fWAR": 0, "Total bWAR": 0}
    for p in players[e][0]: 
        totals[e]["Total fWAR"] += players[e][0][p]["fWAR"]
        totals[e]["Total bWAR"] += players[e][0][p]["bWAR"]
        if "PA" in players[e][0][p]:
            totals[e]["Total PA"] += players[e][0][p]["PA"]
        if "IP" in players[e][0][p]:
            IP = players[e][0][p]["IP"]
            if ".1" in str(IP):
                IP += .23
            elif ".2" in str(IP):
                IP += .47
            totals[e]["Total IP"] += IP
totals
```

```python
# for all the years, add the fWAR, bWAR, PA, and IP and store in dict

totalfwar = 0
totalbwar = 0
totalpa = 0
totalip = 0
for e in totals:
    totalfwar += totals[e]["Total fWAR"]
    totalbwar += totals[e]["Total bWAR"]
    totalpa += totals[e]["Total PA"]
    totalip += totals[e]["Total IP"]
fulltotal = {"Total fWAR": totalfwar, "Total bWAR": totalbwar, "Total PA": totalpa, "Total IP": totalip}
print(fulltotal)
```

```python
# calculate the WAR/player in several varieties, e.g. fWAR per player

fWARbatters = 0
fWARpitchers = 0
bWARpitchers = 0
bWARpitchers = 0
batters = 0
pitchers = 0
totalplayers = 0
for e in players:
    for f in players[e]:
        totalplayers += len(f)
for e in players:
    for p in players[e][0]: 
        if "IP" in players[e][0][p]:
            pitchers += 1
            try: 
                fWARpitchers += players[e][0][p]["fWAR"]
                bWARpitchers += players[e][0][p]["bWAR"]
            except:
                print("")
batters = totalplayers - pitchers
fWARbatters = totalfwar - fWARpitchers
bWARbatters = totalbwar - bWARpitchers
print(totalfwar/totalplayers, totalbwar/totalplayers, fWARbatters/batters, bWARbatters/batters, fWARpitchers/pitchers, bWARpitchers/pitchers) 
```

```python
# all the players with WAR > 1 in a given year

for e in players:
    for f in players[e]:
        for player in f:
            if f[player]["fWAR"] > 1:
                print(player, e, "fWAR", f[player]["fWAR"])
            if f[player]["bWAR"] > 1:
                print(player, e, "bWAR", f[player]["bWAR"])
```

```python
# store in json

with open('players.json', 'w') as outfile:
    json.dump(players, outfile)
with open('totalbyyear.json', 'w') as outfile:
    json.dump(totals, outfile)
with open('fulltotal.json', 'w') as outfile:
    json.dump(fulltotal, outfile)
```
