```python
import pandas as pd
```

```python
b2017 = pd.read_csv("2017 bleaders.csv")
b2017.head()
```

```python
p2017 = pd.read_csv("2017 pleaders.csv")
p2017.head()
```

```python
f2016 = pd.read_csv("2016.csv")
f2016 = f2016[f2016["PLAYER"]!="PLAYER"]
f2016.head()
```

```python
ls = {}
tot2017 = 0
totIP = 0
totPA = 0
for index1, row1 in f2016.iterrows():
    fa = row1["PLAYER"]
#     ls[fa] = 0
    found = False
    search = b2017
#     n = findplayer(fa, f, s)
    if "P" in row1["POS"]:
        search = p2017
    else:
        search = b2017
    for index2, row2 in search.iterrows():
        if fa == row2["Name"]:
            war = row2["WAR"]
            if "P" in row1["POS"]:
                IP = row2["IP"]
                ls[fa] = {"WAR":war, "IP":IP}
                totIP += IP
            else:
                PA = row2["PA"]
                ls[fa] = {"WAR":war, "PA":PA}
                totPA += PA
            tot2017 += war
            found = True
        if found == True:
            break
```
